```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/breast_cancer_survival[1].csv")
df=pd.DataFrame(data)
df.head()



```

```python
df.drop(['Date_of_Surgery','Date_of_Last_Visit'],axis=1,inplace=True)
encoder=LabelEncoder();
obj=df.dtypes[df.dtypes=="object"]
obj


```

```python
for i in obj.index:
  df[i]=encoder.fit_transform(df[i])
df
```

```python
x=df.drop(["Patient_Status"],axis=1)
y=df["Patient_Status"]
```

```python
svm=SVC(kernel="linear",C=0.1)
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
svm.fit(x_train,y_train)
knn.fit(x_train,y_train)
```

```python
y_pred_svm=svm.predict(x_test)
y_pred_knn=svm.predict(x_test)
```

```python
accuracy_svm=accuracy_score(y_test,y_pred_svm)
accuracy_knn=accuracy_score(y_test,y_pred_knn)
print("Accuracy of SVM:",accuracy_svm)
print("Accuracy of KNN:",accuracy_knn)
```
